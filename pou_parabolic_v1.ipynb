```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from operator import ifloordiv
from re import T
def draw_trayectory (u,theta):
  theta = np.radians(theta)
  t_flightx = M/(u*np.cos(theta))
  t_flighty = 2*u*np.sin(theta)/g
  intervaly = np.arange (0, t_flighty,0.001)
  x = []
  y = []
  for t in intervaly:
      Y=u*np.sin(theta)*t-0.5*g*t*t
      X=u*np.cos(theta)*t
      if t < t_flightx:
        y.append(Y)
        x.append(X)
      elif t < 2*t_flightx:
        y.append(Y)
        x.append(2*M-X)
      elif t < 3*t_flightx:
        y.append(Y)
        x.append(X-2*M)
      elif t < 4*t_flightx:
        y.append(Y)
        x.append(4*M-X)
      elif t < 5*t_flightx:
        y.append(Y)
        x.append(X-4*M)
      elif t < 6*t_flightx:
        y.append(Y)
        x.append(6*M-X)
      elif t < 7*t_flightx:
        y.append(Y)
        x.append(X-6*M)
      elif t < 8*t_flightx:
        y.append(Y)
        x.append(8*M-X)
      elif t < 9*t_flightx:
        y.append(Y)
        x.append(X-8*M)
      elif t < 10*t_flightx:
        y.append(Y)
        x.append(10*M-X)
      elif t < 11*t_flightx:
        y.append(Y)
        x.append(X-10*M)


  plt.plot(x,y)
  plt.xlabel('Distance (m)')
  plt.ylabel('Height (m)')
  plt.title('Projectile motion')
```
